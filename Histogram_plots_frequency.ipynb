```python
#combining BEGE, COAX, ICPC and PPC data for each run 
DICT_1 = pd.concat([BN1, CN1, IN1, PN1], ignore_index=True)
DICT_2 = pd.concat([BN2, CN2, IN2, PN2], ignore_index=True)
DICT_3 = pd.concat([BN3, CN3, IN3, PN3], ignore_index=True)
DICT_4 = pd.concat([BN4, CN4, IN4, PN4], ignore_index=True)
DICT_5 = pd.concat([BN5, CN5, IN5, PN5], ignore_index=True)
DICT_6 = pd.concat([BN6, CN6, IN6, PN6], ignore_index=True)
DICT_7 = pd.concat([BN7, CN7, IN7, PN7], ignore_index=True)
DICT_8 = pd.concat([BN8, CN8, IN8, PN8], ignore_index=True)
DICT_9 = pd.concat([BN9, CN9, IN9, PN9], ignore_index=True)
DICT_10 = pd.concat([BN10, CN10, IN10, PN10], ignore_index=True)
DICT_11 = pd.concat([BN11, CN11, IN11, PN11], ignore_index=True)
DICT_12 = pd.concat([BN12, CN12, IN12, PN12], ignore_index=True)
#combining all runs to form one data set 
ALL_EVENTS = pd.concat([DICT_1, DICT_2, DICT_3, DICT_4,DICT_5,DICT_6,DICT_7,DICT_8,DICT_9,DICT_10,DICT_11,DICT_12], ignore_index=True)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#grouping data by detector name 
grouped_data = ALL_EVENTS.groupby('Detector Name')['Number of discharge events'].apply(list)

number = 0 #counter 
means = [] #empty list of average number of events 
for detector_name, numbers in grouped_data.items():
    number = number + 1
    average = np.mean(numbers)
    print(average)
    means.append(average)
    print(f"Detector: {detector_name}")
    print(f"Numbers: {numbers}")
```

```python
unique_elements, counts = np.unique(means, return_counts=True)#unique mean value of events 

for element, count in zip(unique_elements, counts):
    print(f"{element}: {count} occurrences") #number of detectors with a certain average number of events 

plt.pie(counts, labels=unique_elements, autopct='%1.1f%%', startangle=90)
plt.axis('equal') 
plt.show()
```

```python
sum_all = ALL_EVENTS['Number of discharge events'].sum() #sum of the number of events 

total_time = 25*12 #in milli seconds 

# Add a new column by dividing the existing column by the sum (this will be the frequency of events)
ALL_EVENTS['New Column'] = ALL_EVENTS['Number of discharge events'] / total_time

print(ALL_EVENTS)
```

```python
#histogram for the number of discharge events for all events 
selected_column = 'Number of discharge events' 
plt.hist(ALL_EVENTS[selected_column], bins=15, color='blue', edgecolor='black')

plt.xlabel('Number of discharge events')
plt.ylabel('Count')
plt.title(selected_column + ' in all detectors ')
plt.show()
```

```python
def plot_hist(dictionary,tit):
    """ Plots a histogram for the number of discharge events.
    Inputs:
    dictionary - containing the data for the histogram
    tit - title for the histogram
    """
    selected_column = 'Number of discharge events'

    plt.hist(dictionary[selected_column], bins=15, color='blue', edgecolor='black')
    plt.xlabel('Number of discharge events')
    plt.ylabel('Count')
    plt.title( selected_column + ' in ' + tit + ' detectors')
    
```

```python
#dictionaries for BEGE, COAX, PPC an ICPC detectors 
B = pd.concat([BN1, BN2, BN3,BN4,BN5,BN6,BN7,BN8,BN9,BN10,BN11,BN12], ignore_index=True)
C = pd.concat([CN1, CN2, CN3,CN4,CN5,CN6,CN7,CN8,CN9,CN10,CN11,CN12], ignore_index=True)
P = pd.concat([PN1, PN2, PN3,PN4,PN5,PN6,PN7,PN8,PN9,PN10,PN11,PN12], ignore_index=True)
I = pd.concat([IN1, IN2, IN3,IN4,IN5,IN6,IN7,IN8,IN9,IN10,IN11,IN12], ignore_index=True)
```

```python
sum_all = B['Number of discharge events'].sum() #sum of all bege events

total_time = 25*12 #in milli seconds 

B['New Column'] = B['Number of discharge events'] / total_time #frequency of events 

selected_column = 'New Column'

plt.hist(B[selected_column], bins=15, color='blue', edgecolor='black')
plt.xlabel('Frequency (mHz)')
plt.ylabel('Count')
plt.title('Frequency of all discharge events in bege detectors')
plt.show()
```

```python
sum_all = P['Number of discharge events'].sum() #sum of all ppc events

total_time = 25*12 #in milli seconds 

P['New Column'] = P['Number of discharge events'] / total_time #frequency of events 


selected_column = 'New Column' 

# Create a histogram
plt.hist(P[selected_column], bins=15, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Frequency (mHz)')
plt.ylabel('Count')
plt.title('Frequency of all discharge events in ppc detectors')

# Show the plot
plt.show()
```

```python
sum_all = I['Number of discharge events'].sum() #sum of all icpc events

total_time = 25*12 #in milli seconds 

I['New Column'] = I['Number of discharge events'] / total_time


selected_column = 'New Column' 

plt.hist(I[selected_column], bins=15, color='blue', edgecolor='black')
plt.xlabel('Frequency (mHz)')
plt.ylabel('Count')
plt.title('Frequency of all discharge events in icpc detectors')
plt.show()
```

```python
sum_all = C['Number of discharge events'].sum()#sum of all coax events

total_time = 25*12 #in milli seconds 

C['New Column'] = C['Number of discharge events'] / total_time


selected_column = 'New Column'

plt.hist(C[selected_column], bins=15, color='blue', edgecolor='black')
bin_edges, _ = np.histogram(C[selected_column], bins=10)
plt.xlabel('Frequency (mHz)')
plt.ylabel('Count')
plt.title('Frequency of all discharge events in coax detectors')
plt.show()
```

```python
#number of discharge events, detector type 
types = [B,C,P,I]
tit = [" bege"," coax"," ppc"," icpc"]
 
for j, i in zip(tit, types):
    plot_hist(i, j)
    plt.show()
    plt.clf()
```

```python
#number of discharge events, run type 
dicts = [DICT_1,DICT_2,DICT_3,DICT_4,DICT_5,DICT_6,DICT_7,DICT_8,DICT_9,DICT_10,DICT_11,DICT_12]
tit = [" run 1"," run 2"," run 3"," run 4"," run 5", " run 6"," run 7"," run 8"," run 9"," run 10"," run 11"," run 12"]
 
for j, i in zip(tit, dicts):
    plot_hist(i, j)
    plt.show()
    plt.clf()
```
